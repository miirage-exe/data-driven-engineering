```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

print(df.head())
print(df.shape)
# print(sorted(df["airfoil"].unique()))

# -----------------------------
# PCA FUNCTION
# -----------------------------
def my_PCA(D):
    D = np.asarray(D, dtype=np.float64)

    mu = np.mean(D, axis=0)
    sigma = np.std(D, axis=0)
    sigma[sigma == 0] = 1.0

    D0 = (D - mu) / sigma

    C = np.cov(D0, rowvar=False)

    L, A = np.linalg.eigh(C)

    idx = np.argsort(L)[::-1]
    L = L[idx]
    A = A[:, idx]

    Z = D0 @ A

    return Z, L, A, mu, sigma


# -----------------------------
# Preprocessing
# -----------------------------
df_encoded = pd.get_dummies(df, columns=["airfoil"])
df_encoded = df_encoded.apply(pd.to_numeric, errors='coerce')
df_encoded = df_encoded.dropna()

X = df_encoded.to_numpy(dtype=np.float64)
print(X.shape)
print(X[1])
feature_names = df_encoded.columns


# -----------------------------
# PCA
# -----------------------------
Z, L, A, mu, sigma = my_PCA(X)

explained_variance_ratio = L / np.sum(L)
cumulative_variance = np.cumsum(explained_variance_ratio)


# -----------------------------
# Reconstruction (k components)
# -----------------------------
k = 15
k = min(k, Z.shape[1])

Z_k = Z[:, :k]
A_k = A[:, :k]

X_recon_std = Z_k @ A_k.T
X_recon = X_recon_std * sigma + mu
```

```python
# -----------------------------
# PCA plots
# -----------------------------
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(L)+1), explained_variance_ratio, alpha=0.5, label="Individual variance")
plt.step(range(1, len(L)+1), cumulative_variance, where='mid', color='red', label="Cumulative variance")
plt.xlabel("Principal Component Index")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA Explained Variance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -----------------------------
# PCA projection colored by airfoil
# -----------------------------
airfoil_labels = df.loc[df_encoded.index, "airfoil"]

plt.figure(figsize=(8, 6))
for name in airfoil_labels.unique():
    idx = airfoil_labels == name
    plt.scatter(Z[idx, 0], Z[idx, 1], alpha=0.5, label=name)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Space by Airfoil")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# RECONSTRUCTION ERROR

# -----------------------------
# OUTLIER DETECTION 1
# Reconstruction error (BEST METHOD)
# -----------------------------
recon_error = np.linalg.norm(X - X_recon, axis=1)
threshold_recon = np.percentile(recon_error, 95)
outliers_recon = recon_error > threshold_recon

# -----------------------------
# VISUALIZATION: Reconstruction outliers
# -----------------------------
plt.figure(figsize=(8, 6))

plt.scatter(Z[~outliers_recon, 0], Z[~outliers_recon, 1],
            alpha=0.5, label="normal")

plt.scatter(Z[outliers_recon, 0], Z[outliers_recon, 1],
            color="red", label="reconstruction outliers")

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Outliers - Reconstruction Error")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Mahalanobis distance

# -----------------------------
# OUTLIER DETECTION 2
# PCA / Mahalanobis-style
# -----------------------------
eigenvalues = L
dist_M = np.sum(Z**2 / eigenvalues, axis=1)

threshold_maha = np.percentile(dist_M, 95)
outliers_maha = dist_M > threshold_maha

# -----------------------------
# VISUALIZATION: Mahalanobis outliers
# -----------------------------
plt.figure(figsize=(8, 6))

plt.scatter(Z[~outliers_maha, 0], Z[~outliers_maha, 1],
            alpha=0.5, label="normal")

plt.scatter(Z[outliers_maha, 0], Z[outliers_maha, 1],
            color="orange", label="PCA-space outliers")

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Outliers - PCA Distance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -----------------------------
# Store results
# -----------------------------
df_final = df.loc[df_encoded.index].copy()
df_final["recon_error"] = recon_error
df_final["outlier_recon"] = outliers_recon
df_final["outlier_maha"] = outliers_maha


print("\nOutliers (reconstruction):", np.sum(outliers_recon))
print("Outliers (PCA distance):", np.sum(outliers_maha))

# -----------------------------
# OUTLIER CURVES PLOT: CL & CD vs alpha
# -----------------------------

airfoils = df_final["airfoil"].values
alpha = df_final["alpha"].values
CL = df_final["CL"].values
CD = df_final["CD"].values

mask_out = outliers_recon   # or outliers_maha

unique_airfoils = np.unique(airfoils)

plt.figure(figsize=(15, 5))

# -----------------------------
# CL(alpha)
# -----------------------------
plt.subplot(1, 2, 1)

for af in unique_airfoils:
    idx = airfoils == af

    # sort by alpha for clean curves
    sort_idx = np.argsort(alpha[idx])

    a_sorted = alpha[idx][sort_idx]
    cl_sorted = CL[idx][sort_idx]
    out_sorted = mask_out[idx][sort_idx]

    # full curve (black line)
    plt.plot(a_sorted, cl_sorted, color="black", alpha=0.3)

    # normal points
    plt.scatter(a_sorted[~out_sorted],
                cl_sorted[~out_sorted],
                color="black", s=10)

    # outliers (red points + red line segment)
    plt.scatter(a_sorted[out_sorted],
                cl_sorted[out_sorted],
                color="red", s=30)

plt.xlabel("Alpha")
plt.ylabel("CL")
plt.title("CL vs Alpha (Outliers Highlighted)")
plt.grid(True)


# -----------------------------
# CD(alpha)
# -----------------------------
plt.subplot(1, 2, 2)

for af in unique_airfoils:
    idx = airfoils == af
    sort_idx = np.argsort(alpha[idx])

    a_sorted = alpha[idx][sort_idx]
    cd_sorted = CD[idx][sort_idx]
    out_sorted = mask_out[idx][sort_idx]

    # full curve
    plt.plot(a_sorted, cd_sorted, color="black", alpha=0.3)

    # normal points
    plt.scatter(a_sorted[~out_sorted],
                cd_sorted[~out_sorted],
                color="black", s=10)

    # outliers
    plt.scatter(a_sorted[out_sorted],
                cd_sorted[out_sorted],
                color="red", s=30)

plt.xlabel("Alpha")
plt.ylabel("CD")
plt.title("CD vs Alpha (Outliers Highlighted)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# step 1 extract features

def extract_naca_features(name):
    # Remove the "NACA" prefix and any spaces, then convert to uppercase.
    # e.g. "naca 2412" → "2412", "NACA0013" → "0013"
    digits = str(name).upper().replace("NACA", "").replace(" ", "").strip()

    # If what remains is not a pure number string, we cannot parse it → return NaN.
    # NaN rows will be removed later by dropna().
    if not digits.isdigit():
        return pd.Series({"camber": np.nan, "camber_pos": np.nan, "thickness": np.nan, "thickness_position": np.nan})

    if len(digits) == 4:
        # Extract the first two digits individually as integers.
        d1, d2 = int(digits[0]), int(digits[1])

        # Physical decoding of the 4-digit NACA convention:
        return pd.Series({
            "camber":     d1 / 100.0,        # e.g. 2 → 0.02 (2% chord)
            "camber_pos": d2 / 10.0,          # e.g. 4 → 0.4  (40% chord)
            "thickness":  int(digits[2:]) / 100.0,  # e.g. "12" → 0.12 (12% chord)
            "thickness_position": 29.03

        })

    elif len(digits) == 5:
        # 5-digit NACA series has a different (more complex) convention.
        # This is an approximation — sufficient for our purposes.
        d1, d2 = int(digits[0]), int(digits[1])

        return pd.Series({
            "camber":     0.0208,  # approximate series-5 camber
            "camber_pos": 0.2042,
            "thickness":  int(digits[3:]) / 100.0,
            "thickness_position": 29.73
        })

    # Any other length (e.g. 3-digit, 6-digit) → invalid → return NaN.
    return pd.Series({"camber": np.nan, "camber_pos": np.nan, "thickness": np.nan, "thickness_position": np.nan})

df[["camber","camber_pos","thickness","thickness_position"]] = df["airfoil"].apply(extract_naca_features)

# Remove any rows where feature extraction failed (returned NaN).
df = df.dropna()

# print("new excel with extrated parameters")
#print(df[2000:2050])
```
